```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def collatz(n: int) -> int:
    assert n > 0
    if n % 2 == 0:
        return(n // 2)
    else:
        return (3*n + 1)
```

```python
def collatz_conjencture(n: int) -> np.ndarray:
    l = [n]
    while l[-1] != 1:
        l.append(collatz(l[-1]))
    return np.array(l)
```

```python
Y = collatz_conjencture(19)
X = np.arange(0, len(Y))
```

```python
fig, ax = plt.subplots()
ax.plot(X, Y)
```

```python
def longest_collatz_sequence(max_val: int):

    max_len = 1
    input_val = 1

    for num in range(2, max_val):

        temp_len = len(collatz_conjencture(num))
        if temp_len > max_len:
            input_val = num
            max_len = temp_len
        
    return input_val, max_len
```

```python
%% timeit
longest_collatz_sequence(100000)
```
